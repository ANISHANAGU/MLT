```python
# ===========================
# STEP 1: IMPORT LIBRARIES
# ===========================

import pandas as pd
import numpy as np
import time

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, LeaveOneOut, cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score

import matplotlib.pyplot as plt

from google.colab import files
```

```python
# ===========================
# STEP 2: UPLOAD DATASET
# ===========================

uploaded = files.upload()
```

```python
# ===========================
# STEP 3: LOAD DATASET
# ===========================

df = pd.read_csv("FCI_PDS_Demand_Dataset_India_2019_2024.csv")

print("Dataset Shape:", df.shape)
print(df.head())

print("\nMissing values BEFORE preprocessing:\n", df.isnull().sum())
```

```python
# ===========================
# STEP 4: HANDLE MISSING VALUES
# ===========================

num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values AFTER preprocessing:\n", df.isnull().sum())
```

```python
# ===========================
# STEP 5: ENCODE CATEGORICAL DATA
# ===========================

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# ===========================
# STEP 6: FEATURES & TARGET
# ===========================

y = df['total_foodgrain_distributed_tonnes']
X = df.drop('total_foodgrain_distributed_tonnes', axis=1)
```

```python
# ===========================
# STEP 7: NORMALIZATION
# ===========================

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# ===========================
# STEP 8: TARGET CLASSIFICATION
# ===========================

# Convert target variable into 3 classes
y = pd.qcut(df['total_foodgrain_distributed_tonnes'], q=3, labels=[0,1,2])

# Convert categorical to integer
y = y.astype(int)

print("Target Class Distribution:")
print(y.value_counts())
```

```python
# ===========================
# STEP 9: TRAIN TEST SPLIT
# ===========================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
```

```python
# ===========================
# STEP 10: DEFINE MODELS
# ===========================

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC()
}
```

```python
# ===========================
# STEP 11: HOLDOUT RESULTS
# ===========================

print("\n===== HOLDOUT RESULTS =====")

holdout_results = []

for name, model in models.items():

    start = time.time()

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    end = time.time()

    acc = accuracy_score(y_test, preds)
    precision = precision_score(y_test, preds, average='weighted')
    recall = recall_score(y_test, preds, average='weighted')
    f1 = f1_score(y_test, preds, average='weighted')
    f2 = fbeta_score(y_test, preds, beta=2, average='weighted')

    holdout_results.append([name, acc, precision, recall, f1, f2, end-start])

holdout_df = pd.DataFrame(
    holdout_results,
    columns=["Model","Accuracy","Precision","Recall","F1 Score","F2 Score","Training Time"]
)

print(holdout_df)
```

```python
# ===========================
# STEP 12: K-FOLD VALIDATION
# ===========================

kf = KFold(n_splits=5)

kfold_results = []

for name, model in models.items():

    scores = cross_val_score(model, X_scaled, y, cv=kf)

    kfold_results.append([name, scores.mean(), scores.std()])

kfold_df = pd.DataFrame(kfold_results, columns=["Model","Mean Accuracy","Std Dev"])

print(kfold_df)
```

```python
# ===========================
# STEP 13: STRATIFIED K-FOLD
# ===========================

skf = StratifiedKFold(n_splits=5)

strat_results = []

for name, model in models.items():

    scores = cross_val_score(model, X_scaled, y, cv=skf)

    strat_results.append([name, scores.mean(), scores.std()])

strat_df = pd.DataFrame(strat_results, columns=["Model","Mean Accuracy","Std Dev"])

print(strat_df)
```

```python
# ===========================
# STEP 14: LEAVE ONE OUT
# ===========================

loo = LeaveOneOut()

loo_results = []

for name, model in models.items():

    scores = cross_val_score(model, X_scaled, y, cv=loo)

    loo_results.append([name, scores.mean()])

loo_df = pd.DataFrame(loo_results, columns=["Model","LOO Accuracy"])

print(loo_df)
```

```python
final_table = holdout_df[['Model','Accuracy','Precision','Recall','F1 Score','F2 Score']]
print(final_table)

print("\nALL CLASSIFICATION IMPLEMENTATION COMPLETED SUCCESSFULLY!")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

ann_model = Sequential()
ann_model.add(Input(shape=(X_train.shape[1],)))
ann_model.add(Dense(16, activation='relu'))
ann_model.add(Dense(8, activation='relu'))
ann_model.add(Dense(3, activation='softmax'))

ann_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

ann_model.fit(X_train, y_train, epochs=20, batch_size=10, verbose=1)

loss, accuracy = ann_model.evaluate(X_test, y_test, verbose=0)
print("ANN Accuracy:", accuracy)
```

```python
# ===========================
# STEP 15: LABEL DEMAND CATEGORY
# ===========================

def demand_label(val):
    if val == 0:
        return "Low Demand"
    elif val == 1:
        return "Medium Demand"
    else:
        return "High Demand"

y_labels = y.apply(demand_label)

print(y_labels.value_counts())
```

```python
# ===========================
# STEP 16: RULE-BASED RECOMMENDER SYSTEM
# ===========================

def recommend_pds(demand_class):

    if demand_class == 2:   # High Demand
        return [
            "Increase foodgrain allocation immediately",
            "Ensure buffer stock availability",
            "Improve transportation logistics",
            "Monitor supply chain continuously"
        ]

    elif demand_class == 1:   # Medium Demand
        return [
            "Maintain current supply levels",
            "Monitor demand fluctuations",
            "Optimize storage and distribution",
            "Plan for seasonal demand increase"
        ]

    else:   # Low Demand
        return [
            "Reduce excess allocation",
            "Redistribute surplus stock to high-demand regions",
            "Optimize warehouse usage",
            "Avoid overstocking"
        ]
```

```python
# ===========================
# STEP 17: APPLY RECOMMENDER
# ===========================

# Use best model (Random Forest usually performs well)
best_model = RandomForestClassifier()
best_model.fit(X_train, y_train)

predictions = best_model.predict(X_test)

# Create result dataframe
results_df = X_test.copy()
results_df['Actual_Class'] = y_test.values
results_df['Predicted_Class'] = predictions

# Convert to labels
results_df['Demand_Category'] = results_df['Predicted_Class'].apply(demand_label)

# Generate recommendations
results_df['Recommendations'] = results_df['Predicted_Class'].apply(recommend_pds)

print(results_df[['Demand_Category','Recommendations']].head())
```

```python
# ===========================
# STEP 18: USER INPUT DEMO
# ===========================

# Example input (same structure as dataset)
sample_input = X_test.iloc[0].values.reshape(1, -1)

pred = best_model.predict(sample_input)[0]

category = demand_label(pred)
recommendation = recommend_pds(pred)

print("\n===== PDS DEMAND RECOMMENDATION SYSTEM =====")
print("Predicted Demand Category:", category)
print("\nRecommendations:")

for r in recommendation:
    print("-", r)
```

```python
# ===========================
# FINAL CLEAN OUTPUT DISPLAY
# ===========================

print("\n" + "="*60)
print("        PDS DEMAND FORECASTING SYSTEM (FCI)")
print("="*60)

# Select best model automatically
best_model_name = holdout_df.sort_values(by='Accuracy', ascending=False).iloc[0]['Model']
best_model = models[best_model_name]
best_model.fit(X_train, y_train)

print("\nBest Model Selected:", best_model_name)

# Take one sample input
sample_input = X_test.iloc[0].values.reshape(1, -1)
pred = best_model.predict(sample_input)[0]

category = demand_label(pred)
recommendation = recommend_pds(pred)

print("\nPredicted Demand Category:", category)

print("\nRecommended Actions:")
for i, r in enumerate(recommendation, 1):
    print(f"{i}. {r}")

print("\n" + "="*60)
print("        SYSTEM OUTPUT GENERATED SUCCESSFULLY")
print("="*60)
```
